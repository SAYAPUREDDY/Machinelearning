#    Ridge regression and lasso

#lets start this by standardizing our data


```python
from sklearn import preprocessing
```

```python
scaler=preprocessing.StandardScaler().fit(x_train)
```

```python
x_train_s=scaler.transform(x_train)
```

```python
x_test_s=scaler.transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
lm_r=Ridge(alpha=0.5)
lm_r.fit(x_train_s,y_train)
```

```python
r2_score(y_test,lm_r.predict(x_test_s))
```

#the summary of the upcoming process is to train our models with multiple values of lamda and we want the best models out of those models which are calculated using multiple lambda values

```python
from sklearn.model_selection import validation_curve
```

```python
validation_curve?
```

```python
param_range=np.logspace(-2,8,100)
```

```python
param_range
```

```python
#now we will train the model for all the lamda values using follwoing syntax
train_scores,test_scores = validation_curve(Ridge(),x_train_s,y_train,param_name="alpha",param_range=param_range,scoring='r2')
```

```python
print(train_scores)
print(test_scores)
```

```python
train_mean=np.mean(train_scores,axis=1)
test_mean=np.mean(test_scores,axis=1)
```

```python
train_mean
```

```python
test_mean
```

```python
#to find the highest r2 values for our model we use this syntax
max(test_mean)
```

```python
sb.jointplot(x=np.log(param_range),y=test_mean)
```

```python
#to find the location of the particular highest point in the plotted graph
np.where(test_mean==max(test_mean))
```

```python
#bu using the above syntax we get the position of the particular alpha position in the array.
#by finding the value of that array postion we get the alpha value where r2 is maximum
param_range[31]
```

```python
lm_r_best=Ridge(alpha= param_range[31])
```

```python
lm_r_best.fit(x_train_s,y_train)
```

```python
r2_score(y_test,lm_r_best.predict(x_test_s))
```

```python
r2_score(y_train,lm_r_best.predict(x_train_s))
```
